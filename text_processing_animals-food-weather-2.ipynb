```python
# Load text data files stored in a two-level folder structure and convert them into a dataframe
import pandas as pd
import numpy as np
from sklearn.datasets import load_files

files = load_files('animals-food-weather', encoding='utf-8')
  # without encoding='utf-8', text will be byte string
files
```

```python
text = files.data
text
```

```python
target = files.target
target
```

```python
textdf = pd.DataFrame({'Text': text, 'Target': target})
textdf['Category'] = textdf['Target'].apply(lambda x: 'animals' if x==0 else ('food' if x==1 else 'weather'))
textdf = textdf.drop('Target', axis='columns')
textdf.info()
textdf
```

```python
# Tokenization: set to lower case and remove stop words,
from sklearn.feature_extraction.text import CountVectorizer
ctvec = CountVectorizer(lowercase=True, stop_words='english')
```

```python
ctvec_sparse = ctvec.fit_transform(textdf.Text)
ctvec_sparse
```

```python
print(ctvec_sparse)   # sparse matrix
```

```python
ctvec_dense = ctvec_sparse.toarray()
ctvec_dense   # dense matrix
```

```python
# show term-document matrix
dictionary = ctvec.get_feature_names()
pd.DataFrame(ctvec_dense, columns=dictionary)
```

```python
# show TFIDF matrix
from sklearn.feature_extraction.text import TfidfTransformer
tfidf = TfidfTransformer()
tfidf
```

```python
tfidf_sparse = tfidf.fit_transform(ctvec_sparse)
tfidf_dense = tfidf_sparse.toarray()
pd.DataFrame(np.round(tfidf_dense, 2), columns=dictionary)
```

```python
# Compute TFIDF matrix directly
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfvec = TfidfVectorizer(lowercase=True, stop_words='english')
tfidfvec_sparse = tfidfvec.fit_transform(textdf.Text)
tfidfvec_dense = tfidfvec_sparse.toarray()
vocabulary = tfidfvec.get_feature_names()
pd.DataFrame(np.round(tfidfvec_dense, 2), columns=vocabulary)
```
